### 17- Find the longest 100 words. List them.

```python
# Import the necessary library 

from itertools import count
```

```python
words_list = twitter_data["all_words"].tolist()

word_list= []
for item in words_list:
    for word in item.split(' '):
        word_list.append(word)

def find_longest_word(given_list, number):
    cnt = count()
    return sorted(given_list, key = lambda w : (len(w), next(cnt)), reverse = True)[:number]


find_longest_word(word_list, 100)
```

### 19- Identify the mostly occurring 50 hashtags.



```python
# Extract the data from text column as pandas data frame

source = twitter_data["text"]

# Convert pandas data frame to text to be able to use in regular expression

source_text = source.to_string()
```

```python
# Hashtags dictionary in decending order 
# Note that all the hashtags used just once

"""
Here, I defined a funtion which 
  i) accepts list as an input
 ii) creates a dictionary with the frequency of the items in a given list
iii) returns the sorted dictionary in descending order
"""
def list_freq(data):
    freq_dict = {}
    for i in data:
        if i in freq_dict:
            freq_dict[i] += 1
        else:
            freq_dict[i] = 1

    descending_dict = {}
    descending_dict_keys = sorted(freq_dict, key=freq_dict.get, reverse=True)  

    for i in descending_dict_keys:
        descending_dict[i] = freq_dict[i]
     
    return descending_dict

hashtags = re.findall(r"[#]\w+", source_text)

dictionary = list_freq(hashtags)

dict_items = dictionary.items()

first_50 = list(dict_items)[:50]

print(first_50)
```

```python
# Import the necessary library 

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
import seaborn as sns
```

```python
"""

Frequency chart of the top 10 words where the stopping words are removed

"""

tweet_text = twitter_data["text"]

def top_10(text, n=None):
    vec1 = CountVectorizer(stop_words=stop_words,
            max_features=2000).fit(text)
    bag_of_words = vec1.transform(text)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in     
                  vec1.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], 
                reverse=True)
    return words_freq[:n]

top_words = top_10(tweet_text, n=10)
top_words_df = pd.DataFrame(top_words)
top_words_df.columns=["Words", "Frequency"]
print(top_words_df)
sns.set(rc={"figure.figsize":(13,8)})
h=sns.barplot(x="Words", y="Frequency", data=top_words_df)
h.set_xticklabels(h.get_xticklabels(), rotation=45)
```
